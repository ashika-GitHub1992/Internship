2)scrape the following details of each product listed in first 3 pages 
of your search results and save it in a data frame and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”.

```python
#importing time
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#defining variable
brand_name=[]
product_name=[]
rating=[]
num_of_ratings=[]
price=[]
ret_exchange=[]
exp_delivery=[]
availability=[]
other_details=[]
product_url=[]
```

```python
time.sleep(1)
for i in range(0,4):  
    #fetch brand name
    try:
        brand=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base']")
        for j in brand:
            brand_name.append(j.text)
    except:
        brand_name.append("-")
        
    #fetch product name
    try:
        product=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
        for p in product:
            product_name.append(p.text)
    except:
        product_name.append("-")
        
    #fetch rating
    urls=[]
    url=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a/span")
    for u in url:
        urls.append(u.get_attribute('href'))
    try:
        ratings=driver.find_element_by_xpath("//i[@class='a-icon a-icon-star-small a-star-small-4 aok-align-bottom']")
        rating.append(ratings.text)
    except:
        rating.append("-")
            
    #fetch number of ratings
    try:
        number=driver.find_elements_by_xpath("//span[@class='a-size-base']")
        for num in number:
            num_of_ratings.append(num.append)
    except:
        num_of_ratings.append("-")
        
    #fetch price
    try:
        watch_price=driver.find_elements_by_xpath("//div[@class='a-row a-size-base a-color-base']/a/span[1]")
        for w in watch_price:
            price.append(w.text)
    except:
        price.append("-")
        
    #fetch return or exchange
    urlr=[]
    urlrs=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for urs in urlrs:
        urlr.append(urs.get_attribute('href'))
    try:
        ret=driver.find_element_by_xpath("//a[@class='nav-a nav-hasImage nav-right']")
        ret_exchange.append(ret.text)
    except:
        ret_exchange.append("-")
            
    #fetch expected delivery date
    try:
        date=driver.find_elements_by_xpath("//span[@class='a-text-bold']")
        for d in date:
            exp_delivery.append(d.text)
    except:
        exp_delivery.append("-")
        
    #fetch other details
    urld=[]
    ul=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a/span")
    for u in ul:
        urld.append(u.get_attribute('href'))
    try:
        details=driver.find_element_by_xpath("//div[@class='a-section a-spacing-medium a-spacing-top-small']/ul")
        other_details.append(details.text)
    except:
        other_details.append("-")
            
    #fetch product url details
    urp=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for pr in urp:
        product_url.append(pr.get_attribute('href'))
        
    nxt_button=driver.find_elements_by_xpath("//li[@class='a-last']/a")
    try:
        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
    
    
```
